```python
import cv2
import numpy as np
import Utils as utils
import os
import mediapipe as mp
import json
mp_holistic = mp.solutions.holistic
mp_drawing = mp.solutions.drawing_utils
mp_hands = mp.solutions.hands
```

```python
path_videos = '../../Videos/VideosLSECv1/'
datos = {}
with mp_holistic.Holistic(
        min_detection_confidence=0.5,
        min_tracking_confidence=0.5
        ) as holistic:
    for folder in os.listdir(path_videos):
        datos_folder = {}
        img_total_count = 0
        print(folder)
        for frase in os.listdir(path_videos+folder):
            datos_frase = {}
            img_frase_count = 0
            print(frase)
            for video in utils.sorted_alphanumeric(os.listdir(path_videos+folder+'/'+frase+'/')):
                datos_video = {}
                img_video_count = 0
                cap = cv2.VideoCapture(path_videos+folder+'/'+frase+'/'+video)
                while(cap.isOpened()):
                    frameId = cap.get(1)
                    ret, frame = cap.read()
                    if (ret != True):
                        break
                    if(not frame.shape == (360, 480, 3)):
                        frame = cv2.resize(frame, (480,360))
                    image, results = utils.mediapipe_detection(frame, holistic)
                    rh_keypoints_x, rh_keypoints_y, lh_keypoints_x, lh_keypoints_y = utils.extract_keypoints_V3(results)
                    if(rh_keypoints_x and rh_keypoints_y):
                        img_video_count += 1
                    if(lh_keypoints_x and lh_keypoints_y):
                        img_video_count += 1
                datos_frase[video] = datos_video
                datos_video['count'] = img_video_count
                img_frase_count += 1
                print(video+' count:'+str(datos_video['count']))
            datos_folder[frase] = datos_frase
            datos_frase['count'] = img_frase_count
            img_total_count += 1
            print('count: '+str(datos_frase['count']))
        datos[folder] = datos_folder
        datos_folder['count'] = img_total_count
        print('count: '+str(datos_folder['count']))
```

```python
for folder in os.listdir(path_videos):
    datos[folder] = 1
```

```python
json.dump(datos, open('Logs/DataSetInformation/dataset.json', 'w'))
```
